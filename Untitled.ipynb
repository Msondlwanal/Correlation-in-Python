```python
# First let's import the packages we will use in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv('movies.csv')
```

# EXPLORING DATA



```python
df.head()
```

```python
df.tail()
```

```python
df1 = df.copy()
```

```python
y = df1['released'].str.replace("\)","").str.split("(",expand=True).rename(columns={0:'released_date',1:'released_place'})
df1.insert(4,'released_date',y['released_date'])
df1.insert(5,'released_place',y['released_place'])
df1.drop(['released'],axis=1,inplace=True)
df1['released_date'] = df1['released_date'].astype(str).astype('datetime64')
```

# CHECK MISSING VALUE 



```python
df1.isna().sum().sort_values(ascending=False)[:-4].reset_index().rename(columns={"index":"columns",0:"Null_Values"})
```

```python
columns_1 = ['rating','company','writer','country','released_place','star']
for i in columns_1:
    df1[i].fillna("Others",inplace=True)

columns_2 = ['runtime','score','votes']
for j in columns_2:
    df1[j].fillna(0,inplace=True) 
columns_3 = ['budget','gross']
for k in columns_3:
    df1[k].fillna(round(np.mean(df1[k])),inplace=True)
```

```python
df1
```

```python
print(df1.dtypes)
```

# CHANGING DATA TYPE


```python
df1['votes'] = df1['votes'].astype('int64')
df1['gross'] = df1['gross'].astype('int64')
df1['budget'] = df1['budget'].astype('int64')
```

```python
pd.set_option('display.max_rows', 20)
```

```python
# Order our Data a little bit to see

df1.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df1.drop_duplicates()
```

```python
sns.regplot(x="score", y="gross", data=df1, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```

```python
correlation_matrix = df1.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df1.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df1.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df1.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
